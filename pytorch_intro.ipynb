### What is Pytorch?
- Numpy 대체 + GPU연산력사용
- 유연성 및 속도제공하는 딥러닝 연구 플랫폼

```
from __future__ import print_function
import torch
```

```
# 5 x 3 matrix생성, 2D Tensor
# 초기화 되지 않는 행렬
x = torch.Tensor(5,3)
print(x) 
```

```
# 무작위로 초기화된 행렬
x = torch.rand(5,3)
print(x)
```

```
x.size()
```

### Operation

```
y = torch.rand(5, 3)
x + y
```

```
torch.add(x, y)
```

```
# 결과tensor를 인자로 제공
result = torch.Tensor(5, 3)
torch.add(x, y, out=result)
print(result)
```

```
# in-place
# y에 x 더하기
y.add_(x)
print(y)
```

### indexing

```
print(y[:,1]) # 2번째 column vector
```

```
# torch.view : size, shape 변경하고 싶을때 사용
x = torch.randn(4, 4)
y = x.view(16)
z = x.view(-1, 8) # (2,8)
print(y, "\n", z)
```
